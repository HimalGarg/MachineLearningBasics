```python
import pandas as pd
```

```python
X = pd.read_csv("train.csv" , index_col = 'Id')
X.dropna(subset = ['SalePrice'] , axis=0, inplace=True)
y = X.SalePrice
X.drop(['SalePrice'] , axis=1, inplace=True)


categorical_var = [c for c in X.columns if X[c].dtype == 'object']
low_categorical_var = [c for c in X.columns if (X[c].dtype == 'object') & (X[c].nunique() < 10)]

high_card_col = list(set(categorical_var) - set(low_categorical_var))

X.drop(high_card_col , axis=1,inplace=True)
```

```python
lcv_nunique = list(map(lambda col: X[col].nunique() , low_categorical_var))
#col_nunique_dict = dict(zip(low_categorical_var , lcv_nunique))
cat_var = list(map(lambda col: list(X[col].unique()) , low_categorical_var))
col_uniqueval_dict = dict(zip(low_categorical_var , cat_var))
```

```python
for col, uniqueval in col_uniqueval_dict.items():
    for val in uniqueval:
        X[f"{col}_{val}"] = (X[col] == val).astype(int)
    
    X.drop(col, axis=1, inplace=True)

X.head()
```
