```python
import pandas as pd
```

```python
X = pd.read_csv("train.csv" , index_col = 'Id')
X.dropna(subset = ['SalePrice'] , axis=0, inplace=True)
y = X.SalePrice
X.drop(['SalePrice'] , axis=1, inplace=True)

categorical_var = [c for c in X.columns if X[c].dtype == 'object']
```

```python
categorical_var[:3]
```

```python
for col in categorical_var:
    unique = X[col].unique()
    arg={var:idx for idx, var in enumerate(unique)}
    X[col] = X[col].map(arg)
```
