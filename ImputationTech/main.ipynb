```python
import pandas as pd
```

```python
X=pd.read_csv('train.csv' , index_col='Id')
X=X.select_dtypes(exclude = ['object'])
```

```python
col_na = [c for c in X.columns if X[c].isnull().any()]
col_na
```

```python
for col in col_na:
    imputation = X[col].mean()
    X[col].fillna(value = imputation , inplace = True)

X
```
